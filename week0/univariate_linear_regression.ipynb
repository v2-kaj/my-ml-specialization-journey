Gradient discent is an algorithm for finding the values of the parameter of the function that best fits our data.

The idea of gradient discent is that for each iteration we take initial values of our parameters and simultaneously
update these we values until they stop changing signficantly between successive iterations at which point we say the function has converged

So the algorith is:
repeat until convergence:{
    w = w - alpha * partial derivativce of the cost function w.r.t w
    b = b - alpha * partial derivative of the cost function w.r.t to b
}

alpha is the learning rate which is a positive value
let's call the partia derivative of the cost function w.r.t w, djdw and the partial derivative of cost function w.r.t b, djdb

djdw = sum all (w*x[i] + b) -y[i]) * x[i] divided by the number of training examples m
djdb = sum all (w*x[i] + b) -y[i]) divided by the number of training examples m

Note that the above formula are derived using calculas

Let's write the function that will compute the djdw and djdb which are called the gradients.

This is the helper function that we need in the gradient decent function.

```python
def compute_gradient(x,w,y,b):
    """
    Computes the gradient for linear regression 
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    Returns
      djdw (scalar): The gradient of the cost w.r.t. the parameters w
      djdb (scalar): The gradient of the cost w.r.t. the parameter b     
     """
    # for number of training examples, m
    m = x.shape[0]
    
    djdw = 0
    djdb = 0

    for i in range(m):
        tmp_djdw = ((w*x[i] + b) - y[i]) * x[i]
        djdw = djdw + tmp_djdw

        tmp_djdb = (w*x[i] + b) - y[i]
        djdb = djdb + tmp_djdb
        
    return djdw/m, djdb/m
        
        
```


Now that we have a function that will calculate the partial derivatives, let's write the gradient descent function. 

remember gradient descent algorithm:

repeat until convergence:{
    w = w - alpha * djdw
    b = b - alpha * djdb
}

```python
def gradient_descent(x, y, w_in, b_in, alpha, compute_gradient, num_iters):
    
    """Performs gradient descent to fit w,b. Updates w,b by taking 
    num_iters gradient steps with learning rate alpha
    
    Args:
      x (ndarray (m,))  : Data, m examples 
      y (ndarray (m,))  : target values
      w_in,b_in (scalar): initial values of model parameters  
      alpha (float):     Learning rate
      num_iters (int):   number of iterations to run gradient descent
      gradient_function: function to call to produce gradient
      
    Returns:
      w (scalar): Updated value of parameter after running gradient descent
      b (scalar): Updated value of parameter after running gradient descent
      """
    
    w = w_in
    b = b_in
    for i in range(num_iters):
        djdw, djdb = compute_gradient(x, w, y, b)
        w = w - alpha * djdw
        b = b - alpha * djdb
    return w, b
    
```

Now how can we know that for our choice of alpha the gradient descent function is converging.
If alpha is too large, our function will diverge,
If alpha is too small the algorith will run too slowly.
